```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt

train = np.loadtxt("click.csv",delimiter=",",skiprows=1)
```

```python
train_x = train[:,0]
train_y = train[:,1]
```

```python
mu = train_x.mean()
sigma = train_x.std()
```

```python
print(mu,sigma)
```

```python
def standardize(x):
    return (x - mu) / sigma
```

```python
train_x = standardize(train_x)
```

```python
def to_matrix(x):
    return np.vstack([np.ones(x.size),x,x**2]).T
```

```python
to_matrix(train_x)
```
